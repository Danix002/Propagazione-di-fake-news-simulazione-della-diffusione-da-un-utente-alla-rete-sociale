# Choroplet

Abbiamo deciso di rappresentare l'andamento dei contagi tramite più mappe coropletiche. 
La prima mappa è stata costruita utilizzando esclusivamente i confini delle regioni ottenuti dai file dell'ISTAT. In questa mappa, l'intensità del colore rappresentante gli 'Infected' aumenta proporzionalmente al numero dei contagi: maggiore è il numero di contagi, più denso è il colore. L'intensità del colore è dunque calcolata sulla base del numero esatto di contagi nella specifica regione. Inoltre, il plot viene salvato sia in un file .png che in un file .svg. Se le cartelle di destinazione non sono già presenti, è necessario crearle: img_output per i file .png e svg_output per i file .svg.

```python
%matplotlib inline
%run Visualization/choroplet_sample_view.py Infected 1 1
```

La seconda mappa è stata realizzata utilizzando il file "italy_complete_view_with_cities.py", che era stato precedentemente impiegato per creare una visualizzazione completa dell'Italia con le città. La mappa così generata è stata poi colorata seguendo gli stessi principi della prima mappa: l'intensità del colore aumenta con il numero dei contagi. Si osserva come le regioni con un maggior numero di città, e quindi presumibilmente più popolose, siano inevitabilmente le più colpite dall'epidemia.
Inoltre, il plot viene salvato sia in un file .png che in un file .svg. Se le cartelle di destinazione non sono già presenti, è necessario crearle: img_output per i file .png e svg_output per i file .svg.
